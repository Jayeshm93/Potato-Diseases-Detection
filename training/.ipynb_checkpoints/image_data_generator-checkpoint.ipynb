```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np

from IPython.display import HTML
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
IMAGE_SIZE = 256

CHANNELS= 3

train_datagen = ImageDataGenerator(
    horizontal_flip = True,
    rescale= 1./255,
    rotation_range= 10
)

train_generator = train_datagen.flow_from_directory(
    'dset/train',
    target_size = (IMAGE_SIZE, IMAGE_SIZE),
    batch_size = 32,
    class_mode = 'sparse',
) 
```

```python
for img, imgla in train_generator:
    print(img.shape)
    print(img[0])
    break
```

```python
validation_datagen = ImageDataGenerator(
    horizontal_flip = True,
    rescale= 1./255,
    rotation_range= 10
)

validation_generator = validation_datagen.flow_from_directory(
    'dset/val',
    target_size = (IMAGE_SIZE, IMAGE_SIZE),
    batch_size = 32,
    class_mode = 'sparse',
) 
```

```python
test_datagen = ImageDataGenerator(
    horizontal_flip = True,
    rescale= 1./255,
    rotation_range= 10
)

test_generator = test_datagen.flow_from_directory(
    'dset/test',
    target_size = (IMAGE_SIZE, IMAGE_SIZE),
    batch_size = 32,
    class_mode = 'sparse',
) 
```

## Train Model

```python
input_shape = (IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
n_classes = 3


model = keras.Sequential([
    keras.layers.InputLayer(input_shape = input_shape),   # given input 
    keras.layers.Conv2D(filters=32, kernel_size = (3,3), activation = 'relu'),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Conv2D(64, (3,3), activation = 'relu'),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Conv2D(64, (3,3), activation = 'relu'),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Conv2D(64, (3,3), activation = 'relu'),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Conv2D(64, (3,3), activation = 'relu'),
    keras.layers.MaxPooling2D((2,2)),
    
    keras.layers.Flatten(),
    keras.layers.Dense(64, activation = 'relu'),
    keras.layers.Dense(n_classes, activation = 'softmax')
])
model.build(input_shape = input_shape)
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam',
              loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = False),
              metrics = ['accuracy']
) 
```

```python
history = model.fit(train_generator, 
                    steps_per_epoch = 47, 
                    batch_size=32,
                    validation_data=validation_generator,
                    validation_steps= 6,
                    verbose=1, epochs = 10)
```

```python
scores = model.evaluate(test_generator)
```

```python
history.history.keys()
```

```python
acc =   history.history['accuracy']
val_acc = history.history['val_accuracy']

loss =   history.history['loss']
val_loss = history.history['val_loss']
```

```python
EPOCHS = 10

plt.figure(figsize=(7,7))
plt.subplot(1,2,1)
plt.plot(range(EPOCHS), acc, label = 'Training Accurancy')
plt.plot(range(EPOCHS), val_acc, label = 'Validation Accurancy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1,2,2)
plt.plot(range(EPOCHS), loss, label = 'Training loss')
plt.plot(range(EPOCHS), val_loss, label = 'Validation loss')
plt.legend(loc='upper right')
plt.title('Training and Validation loss')
plt.show()
```

```python
classes_names = ['Potato___Early_blight', 'Potato___Late_blight', 'Potato___healthy']
```

```python
for image_batch, labels_batch in test_generator:
    first_image = image_batch[0]
    first_label = int(labels_batch[0])
    
    print('first image to predict')
    plt.imshow(first_image)
    print('Actual label:',classes_names[first_label])
    
    batch_prediction = model.predict(image_batch)
    print('Predicted label:',classes_names[np.argmax(batch_prediction[0])])

    break
```
